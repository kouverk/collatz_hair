```python
import matplotlib.pyplot as plt
import math
from tqdm import tqdm

def advance_collatz(i):
    if i % 2 == 0:
        return i // 2
    else:
        return i * 3 + 1

def draw_collatz_hair(ax, n, x_step=1, y_step=1):
    """Draw a single collatz sequence as a 'hair' strand, starting from origin."""
    
    # Build the sequence first
    i = n
    sequence = [n]
    while i != 1:
        i = advance_collatz(i)
        sequence.append(i)
    
    # Reverse it so we start at 1 and end at n
    sequence = sequence[::-1]
    coords_x = []
    coords_y = []
    for i, val in enumerate(sequence, start=1):
        coords_x.append(i)
        coords_y.append(val)

    desired_rgb = (8, 73, 131)
    ax.plot(coords_x, coords_y, color=tuple(x/255 for x in desired_rgb), alpha=0.03, linewidth=1)

def build_collatz_visualization(N): 
    print("Creating figure...")
    fig, ax = plt.subplots(figsize=(12, 16), dpi=500)
    
    for n in tqdm(range(1, N + 1), desc="Drawing hairs"):
        draw_collatz_hair(ax, n)
    
    print("Setting scales...")
    ax.set_xscale('log') 
    ax.set_yscale('log')
    ax.set_aspect('auto')  
    
    print("Tight layout...")
    plt.tight_layout()
    
    print("Rendering to screen...")
    plt.show()
    
    # Uncomment to save high-res version:
    # print("Saving PNG...")
    # plt.savefig(f'collatz_walk_{N}.png', dpi=500, bbox_inches='tight')
    # print("Done!")
    # plt.close()

N = 5000
build_collatz_visualization(N)
```
